# Text-classification

```python
# ! pip install  transformers

from transformers import pipeline
import pandas as pd

text = """When this air fryer works it is great but upon delivery the first day it unfortunately broke.
I selected the return and replace option thinking that when the item was collected I would have a new one delivered.
This was sadly not the case and the day for collection came and went with no update or explanation as to why the collection was not completed.
I contacted Amazon and after going through a few circles trying to actually speak to someone I was informed that the driver couldn't find me.
This I can understand but not them not trying to contact me to help them locate my property.
Finally after three weeks of collection and delivery pending on my item the same item was returned to me "repaired".
This was not the option that I selected as if something is faulty I do not want it back,
I want a new non faulty item sent to me. The airfryer worked well for a few days then the light went out and then it stopped working altogether.
I am now in the pending stage again but this time for a full refund.
The reason I sent it back in the beginning was to avoid it breaking again as I knew it would but here we are anyway.
I just hope I am not waiting an age for the refund to come through so I can purchase an airfryer
elsewhere with a better level of customer service than Amazon provide """
```

```python

classifier = pipeline("text-classification" , model = 'distilbert-base-uncased-finetuned-sst-2-english' )
# Pipeline(name_of_task  , name of the model from hugging face)
# https://huggingface.co/models
```

```python
outputs = classifier(text)
outputs
```

```python

outputs = classifier(text)
pd. DataFrame(outputs)
```

```python
# Get the model's configuration
model_config = classifier.model.config

# Get the labels used by the model
labels = model_config.id2label

# Print the labels
print(labels)
```

# named entity recognition eith pipeline

```python
from transformers import pipeline
import pandas as pd

text = """When this air fryer works it is great but upon delivery the first day it unfortunately broke.
I selected the return and replace option thinking that when the item was collected I would have a new one delivered.
This was sadly not the case and the day for collection came and went with no update or explanation as to why the collection was not completed.
I contacted Amazon and after going through a few circles trying to actually speak to someone I was informed that the driver couldn't find me.
This I can understand but not them not trying to contact me to help them locate my property.
Finally after three weeks of collection and delivery pending on my item the same item was returned to me "repaired".
This was not the option that I selected as if something is faulty I do not want it back,
I want a new non faulty item sent to me. The airfryer worked well for a few days then the light went out and then it stopped working altogether.
I am now in the pending stage again but this time for a full refund.
The reason I sent it back in the beginning was to avoid it breaking again as I knew it would but here we are anyway.
I just hope I am not waiting an age for the refund to come through so I can purchase an airfryer
elsewhere with a better level of customer service than Amazon Egypt provide """
```

```python
# if we didn't spiecify model name it will use the default model for the task
ner_tagger = pipeline("ner",model = "dbmdz/bert-large-cased-finetuned-conll03-english" ,aggregation_strategy="simple" )
outputs = ner_tagger(text)
out = pd.DataFrame(outputs)
out.head()
```

1- The aggregation_strategy parameter is used to determine how overlapping entities are aggregated when extracting named entities from the input text.

2- The aggregation_strategy="simple" means that a simple strategy will be used to handle overlapping entities. When multiple entities overlap in the output, this strategy will likely keep all of them in the result, and you'll get a list of entities for each token.

3- For instance, if the model identifies entities like "ORG" (organization) and "LOC" (location) in overlapping spans, the simple strategy might include both entities in the result, even if they overlap.

4- Other possible values for the aggregation_strategy parameter include:

"first": Keep only the first identified entity when there is an overlap.
"mean": Compute the mean of the scores for the overlapping entities.
"max": Keep the entity with the maximum score when there is an overlap.
The appropriate choice of the aggregation strategy depends on your specific use case and how you want to handle overlapping entities in the NER results. The default value is typically "first". The choice of strategy can affect the precision and recall of the NER system.

# translation text

```python
# ! pip install  transformers
from transformers import pipeline
import pandas as pd

text = """When this air fryer works it is great but upon delivery the first day it unfortunately broke.
I selected the return and replace option. I thougth that when the item was collected I would have a new one delivered, but until now
there is no update or explanation why the collection was not completed. """
```

```python
translator = pipeline("translation_en_to_ar" ,model="marefa-nlp/marefa-mt-en-ar" )
```

```python
outputs = translator(text, clean_up_tokenization_spaces=True, min_length= 65)
print(outputs[0]['translation_text' ])

# clean_up_tokenization_spaces=True: When set to True, this parameter instructs the translation pipeline to clean up tokenization spaces.
# Tokenization is the process of breaking a text into individual units, usually words or subwords.
# Sometimes, there might be extra spaces introduced during tokenization.
# Setting clean_up_tokenization_spaces=True helps to remove or clean up these additional spaces, ensuring a more coherent translation.

#-------------------------
# min_length=65: This parameter sets a minimum length for the output translation. In this case, it is set to 65.
# If the length of the translated text is less than 65 characters, the pipeline may adjust or generate additional content to meet this minimum length requirement.
```
